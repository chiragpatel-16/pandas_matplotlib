```python
import csv
import numpy as np
from matplotlib import pyplot as plt
from collections import Counter
plt.style.use("fivethirtyeight")
with open('langdata.csv') as csv_file:
    csv_reader=csv.DictReader(csv_file)
    row=next(csv_reader)
   ## print(row)
    lang_cnt=Counter()
    for row in csv_reader:
        lang_cnt.update(row['languageworkedwith'].split(';'))
        lang=[]
        popu=[]
        for item in lang_cnt.most_common(15):
            lang.append(item[0])
            popu.append(item[1])
plt.title('Language Respondent')

plt.ylabel('Language')
plt.xlabel('People')

plt.barh(lang,popu)            
plt.tight_layout()           
```
